```python
import pandas as pd
import numpy as np

# Create a sample dataset
data = {
    "id": range(1, 11),
    "name": ["Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Grace", "Hank", "Ivy", "Jack"],
    "age": [25, 30, 35, 40, 22, 29, 31, 27, 23, 26],
    "salary": [50000, 60000, 75000, 80000, 45000, 55000, 62000, 48000, 47000, 51000],
    "department": ["HR", "IT", "Finance", "IT", "HR", "Marketing", "Finance", "IT", "Marketing", "HR"]
}

df = pd.DataFrame(data)
```

```python
df.head()
```

**Question 1:** Create a new column `"age_category"` that labels employees as:
* `"Young"` if age < 30
* `"Mid-Age"` if 30 ≤ age < 40
* `"Senior"` if age ≥ 40

```python
df['age_category'] = df.apply(lambda row: 'Young' if row['age'] < 30 else 'Middle-Aged' if row['age'] < 40 else 'Senior', axis=1)
df.head()
```

**Question 2:** Increase the salary of all employees by 10% using a lambda function.

```python
df['new_salary'] = df['salary'].apply(lambda x: x+x*0.10)
df.head()
```

**Question 3:** Extract the first letter of each employee's department and store it in a new column `"dept_code"`.

```python
df['dept_code'] = df['department'].apply(lambda x: x[0])
```

```python
df.head()
```

**Question 4:** Create a new column `"bonus"` where:
* If department is `"IT"`, bonus = 15% of salary
* If department is `"Finance"`, bonus = 12% of salary
* Otherwise, bonus = 10% of salary

```python
df['bonus'] = df.apply(lambda x: 0.10*x['salary'] if x['department']=='IT' else 0.15*x['salary'] if x['department']=='Finance' else 0.10*x['salary'], axis=1)
df.head()
```

**Question 5:** Mask employee names such that only the first letter is visible, and the rest are replaced with `"*"` (e.g., `"Alice"` → `"A****"`).

```python
df['masked'] = df['name'].apply(lambda x: x[0]+ '*' * (len(x)-1))
df.head()
```

**Question 6:** Flag employees earning more than 60,000 as `"High Earner"`, otherwise `"Regular"`.

```python
df['type'] = df['salary'].apply(lambda x: "high earner" if x>60000 else "low earner")
df.head()
```
