`Object-oriented programming (OOP)` is a programming paradigm that focuses on the organization of code into objects, which encapsulate data and behavior related to that data.


In OOP, an object is an instance of a class, which defines a blueprint for the object's data and behavior. Objects can interact with each other through methods, which are functions defined within a class that manipulate the object's data or call on other objects' methods.


The key features of OOP include:<br><br>
**Inheritance**, which allows classes to inherit properties and behavior from a parent class<br><br> **Polymorphism**, which allows objects to take on different forms depending on the context in which they are used; and <br><br>
**Encapsulation**, which provides a way to hide the implementation details of a class from other parts of the program.


OOP is widely used in modern software development and is supported by many programming languages, including Java, Python, C++, and Ruby. It is a powerful tool for creating modular, reusable, and maintainable code.

Let's take an example of a simple program that calculates the area of a rectangle.

## Without OOP

```python
# Without OOP

length = float(input("Enter the length of the rectangle: "))
width = float(input("Enter the width of the rectangle: "))

area = length * width

print("The area of the rectangle is:", area)
```

## With OOP

```python
# With OOP

class Rectangle:
    
    def __init__(self):
        length = float(input("Enter the length of the rectangle: "))
        width = float(input("Enter the width of the rectangle: "))
        self.length = length
        self.width = width
    
    def area(self):
        return self.length * self.width

    
```

```python
rectangle = Rectangle()
print("The area of the rectangle is:", rectangle.area())
```

In the code without OOP, we simply take the length and width as input, calculate the area using the formula, and output the result. This is a straightforward and simple approach, but it may not be very flexible if we need to perform more complex operations on rectangles or if we need to manage many rectangles. ( In this case, for very simple programs, using OOP might  add unnecessary complexity. However, as the program grows in complexity and more features are added, OOP can help improve the organization and structure of the code.)

In the code with OOP, we define a Rectangle class that has attributes for length and width, and a method to calculate the area. We can then create an instance of the Rectangle class and call its area method to calculate the area of the rectangle. This approach provides more flexibility as we can create multiple instances of the Rectangle class and perform various operations on them.

In summary, using OOP can improve the code's flexibility and extensibility, making it easier to manage and maintain in the long run.

The **`__init__`** method is also known as a **constructor** because it constructs or initializes the object. It takes the **`self`** parameter as the first argument, which refers to the object being created. The **`self`** parameter is used to access the object's attributes and methods.

## Magic Methods or Dunder Methods

In Python, **magic methods** are special methods that have double underscores at the beginning and end of their names. These methods are also called **dunder methods** (short for "double underscore" methods). Magic methods define how Python built-in operations should be carried out on objects of a class.

Some common examples of magic methods include:

`__init__(self, ...)`: The constructor method that is called when an object of a class is created.

`__str__(self)`: The method that returns a string representation of the object.

`__repr__(self)`: The method that returns a string representation of the object that can be used to recreate the object.

`__len__(self)`: The method that returns the length of the object.

`__add__(self, other)`: The method that defines how the + operator should be carried out on objects of the class.

`__eq__(self, other)`: The method that defines how the == operator should be carried out on objects of the class.


```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"{self.name} is {self.age} years old"
```

```python
# you don't have to call the __str__ either
person = Person('bob',24)
print(person)
```

Therefore, anytime you try to `print()` your object, by default it goes to **`__str__`** to figure out how to print.

## Self

In Python, the `self` parameter is used to refer to the instance of a class inside its methods. It is a reference to the object that is being operated on, and allows you to access its properties and methods.

If you observe the `RockPaperScissors` class, we have used `self` in every method by default, also inside the constructor method we have used `self` for all the variables as well, and third, when we call a method using another method within the class.

Any method inside a class can **only** be called and accessed by its object. But when you want to access a method using another method inside the class, you use **self**

```python
# the play method cannot by just access by the contstrutor by just play()
class RockPaperScissors:
    
    # constructor
    def __init__(self):
        self.choices = ["rock", "paper", "scissors"]
        print("Welcome, Let's play Rock-Paper-Scissors!") 
        play() 
        
    def play(self):
        ## some code
        return
```

It has to use **`self.play()`** to talk to the `play()` method. But what we know is any method inside a class can **only** be called and accessed by its object. Then how is this possible.
